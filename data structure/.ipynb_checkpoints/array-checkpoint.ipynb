```python
class Node ():
    def __init__(self, data, nxt):
        self.data = data
        self.nxt = nxt
```

```python
class MyLinkedList():
    def __init__(self):
        self.size = 0
        self.head = Node(None,None)
        pass
    
    def add(self,data):
        new = Node(data, None)
        curr = self.head
        
        while curr.nxt != None :
            curr = curr.nxt
            
        curr.nxt = new
                
        self.size += 1
        
    def insert(self, idx, data):
        count = idx
        curr = self.head
        while count >0: # 삽입하기 하나 전 노드까지
            curr = curr.nxt
            count -= 1
        temp = curr.nxt # 원래 삽입하려고 하던 인덱스에 있던 노드
        curr.nxt = Node(data,curr.nxt)
        self.size += 1
        
    
    def clear(self):
        self.size = 0
        self.head = Node(None,None)
        
        
    def delete(self, val): # delete by value
        prev = self.head
        curr = self.head.nxt
        
        while curr.nxt != None:
            if curr.data == val:
                prev.nxt = curr.nxt
                curr.nxt = None
                self.size -= 1
                return True
            else :
                prev = prev.nxt
                curr = curr.nxt
        return False
    
    def delbyidx(self,idx):
        prev = self.head
        curr = self.head.nxt
        
        count = idx
        while count >0:
            prev = prev.nxt
            curr = curr.nxt
            count -= 1
            
        prev.nxt = curr.nxt
        curr.nxt = None
        self.size -= 1
    
    def get(self,idx):
        if idx >= self.size:
            print('idx is out of range')
            return None
            
        prev = self.head
        curr = self.head.nxt
        
        count = idx
        while count >0:
            prev = prev.nxt
            curr = curr.nxt
            count -= 1
            
        return curr.data
    
    def getidx(self, value):
        idx = 0
        curr = self.head.nxt
        while curr.nxt != None:
            if curr.data == value:
                return idx
            idx += 1
            curr = curr.nxt
        return None
    
    def isEmpty(self):
        if self.head.nxt == None:
            return True
        else:
            return False
    
    
    def contain(self, value):
        curr = self.head.nxt
        while curr.nxt != None:
            if curr.data == value:
                return True
            curr = curr.nxt
        return False
        
        
```

```python
class DoubleLinkNode:
    def __init__(self, data, nxt, prev):
        self.data = data
        self.nxt = nxt
        self.prev = prev    
```

```python
class MyDoubleLinkedArray:
    def __init__(self):
        self.size = 0
        self.head = DoubleLinkNode(None,None,None)
        self.tail = DoubleLinkNode(None,None,None)
        self.head.nxt = self.tail
        self.tail.prev = self.head
        
    def add(self, data):
        curr = self.tail.prev
        curr.nxt = DoubleLinkNode(data,self.tail, curr)
        self.tail.prev = curr.nxt
        self.size +=1
        
    def insert(self, idx, data):
        if idx <= (self.size-1) / 2:
            curr = self.head.nxt
            count = idx
            while count > 0 :
                curr = curr.nxt
                count -= 1
            new = DoubleLinkNode(data,curr,curr.prev)
            curr.prev.nxt = new
            curr.prev = new
            self.size +=1
        else :
            curr = self.tail.prev
            count = self.size -1
            while count - idx > 0:
                curr = curr.prev
                count -= 1
            new = DoubleLinkNode(data,curr,curr.prev)
            curr.prev.nxt = new
            curr.prev = new
            self.size +=1
    
    def clear():
        self.size = 0
        self.head.nxt = self.tail
        self.tail.prev = self.head
        
    def delete(self, value):
        curr1 = self.head.nxt
        curr2 = self.tail.prev
        count = 0
        while count <= self.size/2:
            if (curr1.data == value):
                curr1.prev.nxt = curr1.nxt
                curr1.nxt.prev = curr1.prev
                curr1.nxt = None
                curr1.prev = None
                self.size -= 1
                return True
            elif (curr2.data == value):
                curr2.prev.nxt = curr2.nxt
                curr2.nxt.prev = curr2.prev
                curr2.nxt, curr2.prev = None, None
                self.size -= 1
            else : 
                curr1 = curr1.nxt
                curr2 = curr2.prev
        return False
    
    def delbyidx(self,idx):
        if idx > (self.size-1) :
            print('idx is out of bound')
            return False
        
        if idx <= (self.size -1) /2:
            curr = self.head.nxt
            count = idx
            while count > 0:
                curr = curr.nxt
                count -=1
            curr.prev.nxt = curr.nxt
            curr.nxt.prev = curr.prev
            curr.nxt = None
            curr.prev = None
            self.size -= 1
            return True
        else:
            curr = self.tail.prev
            count = self.size -1
            while count - idx > 0:
                curr = curr.prev
                count -= 1
            curr.prev.nxt = curr.nxt
            curr.nxt.prev = curr.prev
            curr.nxt = None
            curr.prev = None
            self.size -= 1
            return True
        
    def get(self,idx):
        if idx > (self.size-1) :
            print('idx is out of bound')
            return False
        
        if idx <= (self.size -1) /2:
            curr = self.head.nxt
            count = idx
            while count > 0:
                curr = curr.nxt
                count -=1      
            return curr.data
        else:
            curr = self.tail.prev
            count = self.size -1
            while count - idx > 0:
                curr = curr.prev
                count -= 1
            return curr.data
        
    def get_array(self):
        curr = self.head.nxt
        arr = []
        arr.append(curr.data)
        while curr.nxt != self.tail:
            curr = curr.nxt
            arr.append(curr.data)
        return arr
    
    def get_index(self, value):
        curr = self.head.nxt
        idx = 0
        while idx <= self.size:
            if curr.data == value:
                return idx
            else:
                curr = curr.nxt
                idx += 1
        return False
    
    def isEmpty(self):
        if self.head.nxt == self.tail:
            return True
        else : 
            return False
    
    def contain (self, value):
        curr1 = self.head.nxt
        curr2 = self.tail.prev
        count = 0
        
        while count <= self.size / 2:
            if curr1.data == value:
                return True
            elif curr2.data == value :
                return True
            else :
                curr1 = curr1.nxt
                curr2 = curr2.prev
                count += 1
        return False
```
